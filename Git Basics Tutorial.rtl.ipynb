# ~گیت مقدماتی~

### *مقدمه
> گیت یک سیستم کنترل نسخه توزیع شده است که برای ردیابی تغییرات در فایل‌های کامپیوتری و هماهنگی کار بر روی آن فایل‌ها بین چندین نفر استفاده می‌شود. گیت بیشتر برای مدیریت کد منبع در توسعه نرم‌افزار استفاده می‌شود و اجازه می‌دهد تا چندین نسخه از یک پروژه در سراسر مخازن مختلف توزیع شود.

> گیت و گیت‌هاب یکی نیستند!!!

> گیت‌هاب یک سرویس میزبانی برای مخازن گیت است. گیت هاب یک رابط گرافیکی مبتنی بر وب ارائه می‌دهد و شامل ویژگی‌های اضافی مانند پیگیری مشکلات، مدیریت وظایف و ویکی‌های پروژه است.

### *نصب گیت:
##### > ویندوز:
- به وبسایت گیت بروید و آخرین نسخه را دانلود کنید: https://git-scm.com/downloads
- نصاب را اجرا کرده و دستورالعمل‌های ارائه شده در ستاپ نصب گیت را دنبال کنید.
- پس از نصب، با تایپ کردن git --version در خط فرمان یا گیت بش، نصب را تأیید کنید.
##### > مک‌او‌اس:
- اکثر نسخه‌های مک‌او‌اس همراه با گیت پیش‌نصب شده هستند. با اجرای git --version در ترمینال، نصب را تأیید کنید.
- اگر نصب نشده است، آخرین نصاب گیت مک‌او‌اس را دانلود کرده و دستورالعمل‌ها را دنبال کنید.
- به جای آن، می‌توانید از Homebrew با brew install git در ترمینال استفاده کنید.
##### > لینوکس:
- برای سیستم‌های مبتنی بر دبیان مانند اوبونتو، از sudo apt install git-all استفاده کنید.
- برای سیستم‌های مبتنی بر RPM مانند فدورا، از sudo dnf install git-all استفاده کنید.
- اگر به آخرین نسخه نیاز دارید، گزینه‌هایی نیز برای نصب از منبع وجود دارد.

### *دستورات معمول:
- <b>git init:</b> یک مخزن محلی جدید را در سیستم شما میسازد.
- <b>git status:</b> وضعیت دایرکتوری کاری و ناحیه آماده‌سازی را نمایش می‌دهد.
- <b>git add:</b> تغییرات در دایرکتوری کاری را به ناحیه آماده‌سازی اضافه می‌کند.
- <b>git commit:</b> تغییرات را با پیام توصیفی در مخزن ثبت می‌کند.
- <b>git log:</b> نسخه‌های ثبت شده را نمایش می‌دهد، که امکان مشاهده تاریخچه پروژه را فراهم می‌آورد.
- <b>git push:</b> مخزن دور با کامیت‌های انجام شده به صورت محلی را به‌روزرسانی می‌کند.
- <b>git clone:</b> یک کپی از یک مخزن موجود را در یک دایرکتوری جدید ایجاد می‌کند.
- <b>git branch:</b> شاخه‌ها را لیست می‌کند، ایجاد می‌کند یا حذف می‌کند.
- <b>git checkout:</b> بین شاخه‌ها جابجا می‌شود یا فایل‌های سابق را بازیابی می‌کند.
- <b>git pull:</b> تغییرات را از مخزن دور فراخوانی می‌کند و آن‌ها را با شاخه فعلی ادغام می‌کند.
